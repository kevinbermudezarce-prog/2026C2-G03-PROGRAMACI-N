# Problema #1: Calculadora de propina inteligente

## Problema

Se requiere una calculadora que pueda dar el total de la cuenta, el porcentaje que lleva de propina, que indique la propina y el total final.

## Entrada

- Se requiere monto de la cuenta.
- Para este caso se calcula automaticamente la propina, si la cuenta es menor de $50 se calcula un 10% si es mayor se reduce a 8%

## Proceso

- Se solicita al usuario el monto de la cuenta final, con el cual se calcula la propina utilizada, para brindar un monto total a mostrar.

## Salida esperada

- Se busca que al final se muestre el monto en bruto, la propina en base a ese monto y el monto total de la cuenta.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |cuenta de $25    | Que calcule una propina de 10%       |  Monto de la cuenta: $ 25.00 / Propina aplicada: 10% / Monto de propina: $2.50 / --- TOTAL A PAGAR: $27.50 ---             |
| Caso límite | cuenta de $1500 |Que brinde el monto con propina del 8% |Monto de la cuenta: $ 1500.00 / Propina aplicada: 8% / Monto de propina: $120.00 / --- TOTAL A PAGAR: $1620.00 ---                |
| Caso inválido |Cuenta de -$250      |Que no dé ningún resultado y que lo indique por mensaje        | Valide el cobro de la cuenta realizado    |


```python
# Escriba la solución aquí
try:
    #datos de entrada
    monto_cuenta = float(input("Ingrese el monto de la cuenta: ") )
    if monto_cuenta <50:
        porcentaje_propina = 10
    else:
        porcentaje_propina = 8
        
    #Calculos
    propina = monto_cuenta * (porcentaje_propina / 100 )
    total = monto_cuenta + propina
    
    #Resultado a mostrar
    if monto_cuenta <= 0:
        print ("Valide el cobro de la cuenta realizado")
    else: 
        print (f"Monto de la cuenta: ${monto_cuenta: .2f}")
        print (f"Propina aplicada: {porcentaje_propina}%")
        print (f"Monto de propina: ${propina:.2f}")
        print(f"--- TOTAL A PAGAR: ${total:.2f} ---")
        

except ValueError:
    print("Error: Debe ingresar el monto de la cuenta")





```

# Problema #2: Detector de número par/impar

## Problema

| 2 | Detector de número par/impar | Solicita un entero e indica si es par o impar. | `int`, `%`, `if/else` |
Se requiere un detector automatico que determine si el número ingresado es par o impar


## Entrada

- Se solicita el número al usuario

## Proceso

- Se valida si el número es par o impar, se valida para dar un resultado al usuario y opciones si se ingresa de forma erronea.

## Salida esperada

- Se espera que el resultado sea un texto que indique si es par o impar

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 20 |Que dé un resultado par |El número 20 es par |
| Caso límite | -1267 |Que dé un resultado impar  |El número -1,267 es impar |
| Caso inválido |Se deja en blanco | Que devuelta un error indicando que ingrese el número|Error: Debe ingresar un número entero válido |

```python
# Escriba la solución aquí
#Datos
try:
    numero = int(input("Indique el número que requiere revisar: "))

#Calculos
    if numero % 2 == 0:
        print(f"El número {numero:,} es par")
    else:
        print(f"El número {numero:,} es impar")
except ValueError:
    print ("Error: Debe ingresar un número entero válido")
```

# Problema #3: Clasificador de edad

## Problema

| 3 | Clasificador de edad | Solicita edad y clasifica como menor, adulto o adulto mayor. | `int`, `if/elif/else` |
Se requiere poder calcular según la edad ingresada si es un menor, un adulto o un adulto mayor

## Entrada

- Se solicita la edad

## Proceso

- Calcular según la edad indicada en que categoría queda

## Salida esperada

- Indicar el resultado si es menor, adulto o adulto mayor.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |30 |Indicar que es adulto |La persona es adulta |
| Caso límite |90 |Indicar que es adulto mayor |La persona es adulta mayor |
| Caso inválido |-30 |Devolver el proceso y que  se indique una edad valida|Indique una edad valida |


```python
# Datos
try: 
    edad = int(input("Ingrese la edad de la persona: "))
#calculos: 
    if edad < 0:
        print ("Indique una edad valida")
    else: 
        if edad < 18:
            print ("La persona es menor de edad")
        elif edad >= 65:
            print ("La persona es adulta mayor")
        else: 
            print ("La persona es adulta")
        
except ValueError:
    print ("Error: Ingrese un número valido")
```
