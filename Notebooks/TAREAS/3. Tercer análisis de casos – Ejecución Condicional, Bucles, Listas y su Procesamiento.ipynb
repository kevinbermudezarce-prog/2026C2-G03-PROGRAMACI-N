# Caso 2: Asistencia a una actividad cultural

## Problema
Una actividad cultural en Costa Rica registra asistencia por bloques horarios. El programa debe calcular la asistencia promedio y clasificar el nivel de ocupación.

## Entrada / datos
Use esta lista base:

```python
asistencias = [120, 180, 210, 250, 230]
capacidad = 300
```

## Proceso esperado

- Recorrer la lista con un ciclo `for`.
- Calcular asistencia total y promedio.
- Contar cuántos bloques superaron el 70% de capacidad.
- Clasificar el promedio de ocupación:
  - `Alta ocupación` si el promedio es mayor o igual al 80% de la capacidad.
  - `Ocupación media` si el promedio es mayor o igual al 50%.
  - `Baja ocupación` si el promedio es menor al 50%.

## Salida esperada
Debe mostrar total de asistentes, promedio, bloques sobre 70% y estado de ocupación.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|---|---|---|---|
| Caso normal | `[120, 180, 210, 250, 230]`, capacidad `300` | Ocupación media o alta según cálculo | Ocupación media, supera el 50% de la capacidad |
| Caso límite | promedio exactamente 80% de capacidad | Alta ocupación | Alta ocupación, supera el 80% de la capacidad |
| Caso especial | lista con asistencias muy bajas | Baja ocupación | Baja ocupación, no supera el 50% de la capacidad |

````python
# Caso 2 - Escriba su solución aquí

asistencias = [120, 180, 210, 250, 230]
capacidad = 300

# Complete el análisis

def calcular_total(asistencias):
    """Recibo la lista de asistencias, la sumo y retorno el total."""
    return sum(asistencias)

def calcular_promedio(asistencias):
    """Retorna el promedio de las ventas de la semana"""
    return sum(asistencias) / len(asistencias)

def calcular_porcentaje_ocupacion(promedio, capacidad):
    """Muestra el procentaje de ocupación del evento"""
    porcentaje = (promedio / capacidad) * 100
    return porcentaje

def capacidad_bloques(asistencias, capacidad):
    """Se crea la función para poder contabilizar cuantos bloques superan el 70% de la capacidad"""
    contador = 0
    
    for asistencia in asistencias:
        if asistencia >= capacidad * 0.70:
            contador += 1
    return contador
    


def clasificar_ocupacion(promedio, capacidad):
    """Se crea la función para poder clasificar el promedio de ocupación según se requiere para >=80%, >=50% y <50%"""
    if promedio >= capacidad * 0.80:
        return "Alta ocupación, supera el 80% de la capacidad"

    elif promedio >= capacidad * 0.50:
        return "Ocupación media, supera el 50% de la capacidad"

    else:
        return "Baja ocupación, no supera el 50% de la capacidad"



total = calcular_total(asistencias)
promedio = calcular_promedio(asistencias)
bloques = capacidad_bloques(asistencias, capacidad)
estado = clasificar_ocupacion(promedio, capacidad)
porcentaje = calcular_porcentaje_ocupacion(promedio, capacidad)


print("--- Reporte de asistencia a la actividad cultural ---")
print(f"Total de asistentes: {total}")
print(f"Promedio de personas que asistieron a la actividad: {promedio:.2f}")
print(f"Cantidad de bloques que superan el 70% de la capacidad: {bloques}")
print(f"Porcentaje de ocupación: {porcentaje:.2f}%")
print(f"Ocupación según la asistencia al evento:: {estado}")



#Pruebas manuales
print ("\n")

asistencias = [200, 240, 250, 250, 260]
capacidad = 300

# Complete el análisis

def calcular_total(asistencias):
    """Recibo la lista de asistencias, la sumo y retorno el total."""
    return sum(asistencias)

def calcular_promedio(asistencias):
    """Retorna el promedio de las ventas de la semana"""
    return sum(asistencias) / len(asistencias)

def calcular_porcentaje_ocupacion(promedio, capacidad):
    """Muestra el procentaje de ocupación del evento"""
    porcentaje = (promedio / capacidad) * 100
    return porcentaje

def capacidad_bloques(asistencias, capacidad):
    """Se crea la función para poder contabilizar cuantos bloques superan el 70% de la capacidad"""
    contador = 0
    
    for asistencia in asistencias:
        if asistencia >= capacidad * 0.70:
            contador += 1
    return contador
    


def clasificar_ocupacion(promedio, capacidad):
    """Se crea la función para poder clasificar el promedio de ocupación según se requiere para >=80%, >=50% y <50%"""
    if promedio >= capacidad * 0.80:
        return "Alta ocupación, supera el 80% de la capacidad"

    elif promedio >= capacidad * 0.50:
        return "Ocupación media, supera el 50% de la capacidad"

    else:
        return "Baja ocupación, no supera el 50% de la capacidad"



total = calcular_total(asistencias)
promedio = calcular_promedio(asistencias)
bloques = capacidad_bloques(asistencias, capacidad)
estado = clasificar_ocupacion(promedio, capacidad)
porcentaje = calcular_porcentaje_ocupacion(promedio, capacidad)


print("--- Reporte de asistencia a la actividad cultural, segundo escenario ---")
print(f"Total de asistentes: {total}")
print(f"Promedio de personas que asistieron a la actividad: {promedio:.2f}")
print(f"Cantidad de bloques que superan el 70% de la capacidad: {bloques}")
print(f"Porcentaje de ocupación: {porcentaje:.2f}%")
print(f"Ocupación según la asistencia al evento: {estado}")

print ("\n")



print ("\n")

asistencias = [50, 80, 100, 90, 120]
capacidad = 300

# Complete el análisis

def calcular_total(asistencias):
    """Recibo la lista de asistencias, la sumo y retorno el total."""
    return sum(asistencias)

def calcular_promedio(asistencias):
    """Retorna el promedio de las ventas de la semana"""
    return sum(asistencias) / len(asistencias)

def calcular_porcentaje_ocupacion(promedio, capacidad):
    """Muestra el procentaje de ocupación del evento"""
    porcentaje = (promedio / capacidad) * 100
    return porcentaje

def capacidad_bloques(asistencias, capacidad):
    """Se crea la función para poder contabilizar cuantos bloques superan el 70% de la capacidad"""
    contador = 0
    
    for asistencia in asistencias:
        if asistencia >= capacidad * 0.70:
            contador += 1
    return contador
    


def clasificar_ocupacion(promedio, capacidad):
    """Se crea la función para poder clasificar el promedio de ocupación según se requiere para >=80%, >=50% y <50%"""
    if promedio >= capacidad * 0.80:
        return "Alta ocupación, supera el 80% de la capacidad"

    elif promedio >= capacidad * 0.50:
        return "Ocupación media, supera el 50% de la capacidad"

    else:
        return "Baja ocupación, no supera el 50% de la capacidad"



total = calcular_total(asistencias)
promedio = calcular_promedio(asistencias)
bloques = capacidad_bloques(asistencias, capacidad)
estado = clasificar_ocupacion(promedio, capacidad)
porcentaje = calcular_porcentaje_ocupacion(promedio, capacidad)


print("--- Reporte de asistencia a la actividad cultural, tercer escenario ---")
print(f"Total de asistentes: {total}")
print(f"Promedio de personas que asistieron a la actividad: {promedio:.2f}")
print(f"Cantidad de bloques que superan el 70% de la capacidad: {bloques}")
print(f"Porcentaje de ocupación: {porcentaje:.2f}%")
print(f"Ocupación según la asistencia al evento: {estado}")

print ("\n")
````

# Caso 3: Precios de productos de feria

## Problema
Una persona compara precios de un mismo producto en diferentes ferias del agricultor. El programa debe analizar los precios y detectar si hay mucha variación.

## Entrada / datos
Use esta lista base:

```python
producto = "Tomate"
precios = [900, 850, 1000, 950, 1200]
```

## Proceso esperado

- Calcular precio promedio.
- Identificar precio mínimo y máximo.
- Calcular rango de precios: `máximo - mínimo`.
- Clasificar la variación:
  - `Variación alta` si el rango es mayor o igual a 300.
  - `Variación moderada` si el rango es mayor o igual a 150.
  - `Variación baja` si el rango es menor a 150.
- Usar un ciclo para contar cuántos precios están por encima del promedio.

## Salida esperada
Debe mostrar producto, promedio, mínimo, máximo, rango, precios sobre promedio y clasificación.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|---|---|---|---|
| Caso normal | `[900, 850, 1000, 950, 1200]` | Variación alta |Rango entre el precio más alto y el más bajo: ₡350.00, Clasificación del rango: Variación alta, Precios del tomate por encima del promedio: 2 |
| Caso límite | rango exactamente `300` | Variación alta |Rango entre el precio más alto y el más bajo: ₡300.00, Clasificación del rango: Variación alta, Precios del tomate por encima del promedio: 2 |
| Caso especial | todos los precios iguales | Variación baja |Rango entre el precio más alto y el más bajo: ₡0.00, Clasificación del rango: Variación baja, Precios del tomate por encima del promedio: 0 |
